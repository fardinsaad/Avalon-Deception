# Experiments

## Experiment 1: Deception with GPT-4 series (GPT-4o). Used as Ground-truth.

```python
import pandas as pd

# Load the CSV file from the local directory
csv_file_path = 'Deception.csv'  
df = pd.read_csv(csv_file_path)  # Use first row as header
#print(df.head())

# Remove any columns that start with 'Unnamed'
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]

# Ensure the 'Generated_response' column exists and is of type string
colname = 'Generated_response_gpt_4o'
if colname not in df.columns:
    df[colname] = ""  # Initialize with empty strings
else:
    df[colname] = df[colname].astype(str)

num_rows = 1  # Adjust this to process 5 or 10 rows, or any other number
df_subset = df.head(num_rows)
counter = 0
# Loop through each row, read the first and second columns, and generate the response
try:
    for index, row in df.iterrows():
        story = row.iloc[0]
        # print(story)  

        # Initialize the agent
        my_agent = Agent()
        
        # Role and Model
        role = "Good" 
        model = "gpt-4o"

        # Generate the response using the agent and instruction
        response = my_agent.chain_of_thought_prompt(story, role, model)
        
        # Append the generated response directly into the DataFrame
        df.at[index, colname] = response
        counter += 1

        print(f"Processed row {counter}")
        
except Exception as e:
    print(f"Error occurred at row {counter}: {e}")
    # Save what we have so far
    df.to_csv(csv_file_path, index=False)
    raise
finally:
    # Always save at the end
    df.to_csv(csv_file_path, index=False)
    print(f"Completed processing. {counter} rows processed and saved.")
```
